#### Use Pandas dataframe.replace( ) to replace strings, lists, dictionaries, regex and numbers. 

```python
import numpy as np
import pandas as pd
```

```python
dfw = pd.read_csv('washington.csv')
dfw.head(10)
```

```python
dfw.info()
```

```python
# use replace to switch out strings with other strings
dfw.replace('14th & Belmont St NW','road_closed', inplace =True)
dfw
```

#### The map() function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.)

```python
#  mapping values to replace multiple values in the table
mapping = {'1st & K St SE': 1, 'Maine Ave & 7th St SW': 2}
```

```python
dfnew = dfw.replace({'Start Station': mapping, 'End Station': mapping})
dfnew.head(10)
```
