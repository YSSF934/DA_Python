### Example code of how to run SQLite in a Jupyter Notebook

```python
import pandas as pd
import sqlite3
```

First, read dataframe in Pandas

```python
dfp = pd.read_csv('powerplant_data.csv')
```

```python
dfp
```

The next step is to create your SQL database.  This is done with the conn method and naming the database as seen below.

```python
conn = sqlite3.connect('database_power.sqlite')
```

Send your pandas dataframe to your SQL lite database.  This first table is called plant_1.

```python
dfp.to_sql("plant_1",conn)
```

Now you can run SQL queries on the new table.  

```python
df1= pd.read_sql("SELECT * FROM plant_1 ",conn)
df1
```
